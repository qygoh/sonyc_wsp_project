```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

#### From Sid (try 1):

sonycnode-b827eb815321.sonyc

has quite a few construction related complaints on 2016-07-05 from 9pm to 11pm

```python
#reading the pickled file
sensor = pd.read_pickle('sonycnode-b827eb815321.sonyc.p').dropna()

sensor.head()
```

```python
year2016 = sensor[sensor['timestamp'].dt.year == 2016]
year2016.head()
```

```python
test = year2016.resample('1min').mean().reset_index()
```

```python
test[(test.timestamp.dt.month == 7)].dBAS.unique()
```

**Comments: even with resampling, the date of interest returned NAN**


#### From Sid (try 2)

sonycnode-b827ebc6dcc6.sonyc

2017-01-18 from 10pm to 12 pm

```python
sensor2 = pd.read_pickle('sonycnode-b827ebc6dcc6.sonyc.p')

sensor2.head()
```

```python
s2017_1 = sensor2[(sensor2['timestamp'].dt.year == 2017) & (sensor2['timestamp'].dt.month == 1)]
s2017_1['timestamp'].dt.day.unique()
```

**Comments: The original data without resampling has the above dates with data**

Resampled to 1S data below:

```python
# first selected for year 2017 and jan
test2 = sensor2[(sensor2['timestamp'].dt.year == 2017) & (sensor2['timestamp'].dt.month == 1)]
```

```python
# then resampled to 1S
test2_s = test2.resample('1S').mean().reset_index()

test2_s.head()
```

```python
test2_s['timestamp'].dt.day.unique()
```

**Comments: upon resampling to 1S, realized that there is indeed more days appearing**

```python
test2_s[test2_s['timestamp'].dt.day == 6]
```

```python
test2_s[test2_s['timestamp'].dt.day == 7]
```
